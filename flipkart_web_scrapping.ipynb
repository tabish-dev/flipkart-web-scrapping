```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
def get_title(item_parsed):
    try:
        title = item_parsed.find('span', attrs={'class':'B_NuCI'}).text.strip()
    except AttributeError:
        title = ""
    return title


def get_price(item_parsed):
    try:
        price = item_parsed.find('div', attrs={'class':'_30jeq3 _16Jk6d'}).text.strip()
    except AttributeError:
        price = ""
    return price

def get_rating(item_parsed):
    try:
        rating = item_parsed.find('div', attrs={'class':'_3LWZlK'}).text.strip()
    except AtrributeError:
        rating = ""
    return rating

```

```python
if __name__ == '__main__':
    URL = "https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=mobiles&requestId=a7d40707-4204-4596-a6c4-e02fe229947f"
    HEADERS = ({'User-Agent':'',   # Wite your User Agent String here
          'Accept-Language':'en-US, en;q=0.5 '})

    webpage_request = requests.get(URL, headers=HEADERS)
    webpage_parsed = BeautifulSoup(webpage_request.content, 'html.parser')
    webpage_links = webpage_parsed.find_all('a', attrs={'class':'_1fQZEK'})

    links_list = []

    for each_link in webpage_links:
        links_list.append(each_link.get('href'))

    dict = {
        'title':[],
        'price':[],
        'rating':[]
    }

    for each_href in links_list:
        single_item_request = requests.get('https://www.flipkart.com' + each_href, headers=HEADERS)
        single_item_parsed = BeautifulSoup(single_item_request.content, 'html.parser')
        dict['title'].append(get_title(single_item_parsed))
        dict['price'].append(get_price(single_item_parsed))
        dict['rating'].append(get_rating(single_item_parsed))


```

```python
df = pd.DataFrame(dict)
```
